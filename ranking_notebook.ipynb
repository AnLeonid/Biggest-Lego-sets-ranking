```python
import pandas as pd
```

```python
df = pd.read_csv("sets.csv.gz",compression='gzip')
```

```python
df['numname']=df['set_num'].astype(str)+' '+df['name']
```

```python
cols = list(df.columns)
cols = [cols[-1]] + cols[:-1]
df = df[cols]
```

```python
df.head()
```

```python
df.sort_values(by=['num_parts'], ascending=False).head(10)
```

```python
data=pd.DataFrame(data=df, columns=list(df.columns)+sorted(list(set(df.year))+[1951, 1952]))
```

```python
%%time
for i in range(len(data.index)):
    for j in range(6, len(data.columns)):
        if data.columns[j] >= data.iloc[i, 3]:
            data.iloc[i, j] = data.iloc[i, 5]  
```
